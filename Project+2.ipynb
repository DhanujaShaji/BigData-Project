```python
import pandas as pd
import numpy as np
train = pd.read_json("train.json")
train.set_index('listing_id',inplace=True)
```

```python
train.head()
```

```python
all_feat = []
def extractFeatures(lst):
    global all_feat
    for item in lst:
        all_feat.append(item)
    return all_feat

train['features'].apply(extractFeatures);   
```

```python
feat = pd.DataFrame(all_feat,columns = ['feature'])

top25_count = feat["feature"].value_counts().head(25)
```

```python
top25 = list(top25_count.index)
```

```python
def featureBinary(lst,ft):
    if ft in lst:
        return 1
    else:
        return 0
```

```python
for ft in top25:
    train[ft] = train['features'].apply(lambda x:featureBinary(x,ft))
train = train.drop('features',1)
train.head()
```

Extract the class value which is 'interest_level' and store it as a list. To build a basic model taking into consideration only 'features'. Removed all other features and maintained only the binary values features.

```python
labels = train['interest_level']
newTrain = train.drop(['bathrooms','bedrooms','price','building_id','interest_level','created','description','display_address','latitude','longitude','manager_id','photos','street_address'],1)
newTrain.head()
```

```python
labels = labels.tolist()
```

Create an array for values of the features that are used for training.

```python
listing = list(newTrain.index.values)  
values=[]
for l in listing:
    values.append(newTrain.loc[l].tolist())
values = np.array(values)
values.shape
```

Build a Gaussian Naive Bayes Model.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
model = GaussianNB()
```

```python
model.fit(values, labels)
```

Perform 10 fold cross validation

```python
from sklearn.model_selection import cross_val_score
y = labels
scores = cross_val_score(model, values, y, cv=10, scoring='accuracy')
```

Examine the performance

```python
scores.mean()
```
